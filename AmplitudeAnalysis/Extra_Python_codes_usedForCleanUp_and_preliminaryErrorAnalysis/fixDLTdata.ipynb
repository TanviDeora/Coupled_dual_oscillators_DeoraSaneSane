```python
# read data and subtract to invret y axis

filepath = r"D:\analysis - winglength vs frequency\freq vs winglength\raw data\asymmetric wings\fly1/"
xyfiles = glob.glob(filepath + "*xypts.csv")
xyfiles
```

```python
outpath = r"./dataFolder/DLTanalysis/asymWings/"
```

```python
for file in xyfiles:
    print(file)
    df = pd.read_csv(file)
    df.head()
    cols = df.columns[df.columns.str.contains('_Y')]
    
    df_copy = df.copy(deep = True)
    for col in cols:
        if df.loc[:, col].notnull().any():
            bla = 600 - df.loc[:,col]
            df_copy.loc[:,col] = bla
    
    name = os.path.basename(file)

    df_copy.to_csv(outpath + name)
```
