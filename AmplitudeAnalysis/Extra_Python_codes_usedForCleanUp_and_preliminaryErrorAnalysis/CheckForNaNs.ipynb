```python
import glob
import os
import pandas as pd
import numpy as np
```

```python
def getNaNs(dummy, n):
    
    name = []
    frames = []
    points = []

    for index, row in dummy.iterrows():
        for column in row.isnull().index:
            if row.isnull()[column]:
    #             print('found the NaN on %s column and %s index' %(column, index))
                name.append(n)
                frames.append(index)
                points.append(column)
                
    null_df = pd.DataFrame({'name': name,
                       'frameNumber': frames,
                       'Points': points})
    return(null_df)
```

```python
files = glob.glob('../dataFolder/Digitization_Final_Data/*/*/*/*xyzpts.csv', recursive=True)
```

```python
# files
```

```python
Nan_loc = pd.DataFrame(data = None, columns = ['name', 'frameNumber', 'Points'])

for f in files:
    df = pd.read_csv(f)
    name = os.path.basename(f)
    f_null = getNaNs(df, name)
    Nan_loc = Nan_loc.append(f_null)
```

```python
Nan_loc.to_csv('../dataFolder/Digitization_Final_Data/LocationsOfNans.csv')
```

```python
# pick out non contiounous frames!

n = []
f_first = []
f_end = []
for name, subdf in Nan_loc.groupby('name'):
    # find the begining of NaN chunck
    f0=0
    for frame, row in subdf.groupby('frameNumber'):
        f1 = frame
        if f1-f0 != 1:
            n.append(name)
            f_first.append(f1)
        f0=frame
    # find the end of the NaN chunck
    f0=0
    for frame, row in subdf[::-1].groupby('frameNumber', sort = False):
        f1 = frame
        if f1-f0 != -1:
            f_end.append(f1)
        f0=frame
```
